```python
# konl 은 여러가지 형태소를 묶어논 패키지. 
# from konlpy.tag import * 헝태소를 한번에 불러올 패키지. 

from konlpy.tag import * 
```

```python
text = '인생은 모두가 함께하는 여행이다. 매일매일 사는 동안 우리가 할 수 있는 건 최선을 다해 이 멋진 여행을 만끽하는 것이다'
print(text.split(' '))
# 한글은 공백으로 토큰화 하는게 좋은 방법이 아니다.
# 그래서 형태소 분석기를 사용 하는 것이다. 
```

```python
# 코모란 
# 코모란을 사용 하겠다 호출 하는 것. 
from konlpy.tag import Komoran 
# 선언 : 코모란을 사용 하기 위해서는 꼭 선언 해줘야 한다. 
komoran= Komoran() 
```

```python
# 토큰화 : morphs -> 단어들만 뽑아주는 
# morths , nouns(명사), pos -> 이거 세개중에 원하는 것 사용 
komoran_tokens = komoran.morphs (text) 
print(komoran_tokens)
```

```python
# 한나눔 
# 호출을하고, 
from konlpy.tag import Hannanum 
#사용하겠다고 선언을 한다. 
hannanum= Hannanum( )
# (text) 는 타입을 동일하게 사용 한다. 
hannanum_tokens = hannanum.morphs (text)
print(hannanum_tokens) 

# 한나눔 : 굉장히 구체적으로 나눠준다. ex) 멋진 -> 멋지ㄴ 
# 한나눔은 굉장히 구체적으로 나눠주지만, 좀 오래 걸린다. 
```

```python
#Okt (Twitter tokenizer가 v0.5.0.부터 Okt로 변경) 
# Okt -> sns데이터를 가지고 만든 형태소 분석기, 이모티콘도 분석가능 

from konlpy.tag import Okt 
okt= Okt() 
okt_tokens = okt.morphs (text) 
print (okt_tokens) 

```

```python
# Kkma 
from konlpy.tag import Kkma 
kkma= Kkma() 
# morphs, pos, nouns 세개중에 사용 
kkma_tokens = kkma.morphs (text) 
print (kkma_tokens) 
```
