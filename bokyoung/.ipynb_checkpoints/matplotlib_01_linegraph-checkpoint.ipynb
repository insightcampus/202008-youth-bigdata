## III. 선 그래프(line plot)

시계열 데이터와 같이 연속적인 값의 변화와 패턴을 파악하는데 적합하다. 그래프를 그리기 위해 2020년 전세계를 강타했던 코로나 데이터셋을 위주로 그래프를 그려보도록 한다. 

### (1) 전체 소스코드

코로나 데이터셋을 기반으로 데이터 수집부터 시각화까지 전체 소스코드를 보여준 뒤, 다시 설명하도록 한다. 

````python
# 모듈 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# Excel 데이터 데이터프레임으로 변환
covid_19 = pd.read_excel("data\20200406-COVID-19-geographic-disbtribution-worldwide.xlsx")

# KOR 데이터만 추출, val. countryterritoryCode
covid_kor = covid_19[covid_19['countryterritoryCode'] == 'KOR']
covid_kor.set_index('dateRep', inplace=True)

plt.style.use('bmh')

plt.plot(covid_kor.index, covid_kor.cases)
plt.title('Corona Time Series in S.Korea')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend(labels=['Cases'], loc='best')
plt.show()
````

- (4) Matplotlib 그래프 그리기

Matplotlib은 두가지 방식의 API를 제공하고 있다. 
- `Pyplot API` : matplotlib.pyplot 모듈에 함수로 정의 되어 있다. 
- `객체지향 API` : matplotlib이 구현된 객체지향 라이브러리를 직접 활용한다.세밀한 제어를 위해서는 `객체지향 API` 를 사용해야 한다.  



- 선 그래프를 그리는 `plot()` 함수에 입력할 `x, y`축 데이터를 선택한다. 

+ 방법 1: Pyplot API(커맨드 방식)

````python
# X축 index 앞 5개 미리보기

covid_kor.index[:5]
````

````python
#Y축 - 날짜별 개수 앞 5개 미리보기

covid_kor.cases.head()
````

plt.plot([x],y,[fmt],lw,...)

`x`와 `fmt`,`lw`는 생략 가능한 parameter이다. 

- `x`: x축 인덱스
- `y`: y축 데이터
- `fmt`: 그래프 종류와 색 변경
```
fmt = '[marker][line][color]'
```
    - 'r-'은 빨간색 선 그래프
    - 'g--'은 초록색 점선 그래프

    자세한 사항은 [matplotlib](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.plot.html) 참고하면 된다. 
- `lw`: `linewidth` 라고도 쓸 수 있는 parameter로 선 굵기 지정

````python
plt.plot(covid_kor.index, covid_kor.cases,'r-',lw=1) 
````

plt.legend()는 범례를 지정하는 함수로, 여기서는 두 가지 parameter 를 추가 했다. 
- `labels` : 데이터 라벨을 지정한다.
- `loc`: 범례 위치 조정, 여기서는 loc= 'best'로, 디폴트로 적용되는 오른쪽 위에 배치 된다. 

````python
plt.plot(covid_kor.index, covid_kor.cases,'r-',lw=1) 
plt.legend(labels=['Cases'], loc='best')
````

위와 같이 그래프가 출력이 되었다면, 그 다음부터는 그래프를 꾸미는 것이다. 차트의 제목, x축 이름, y축 이름 함수를 활용하여 그래프를 완성한다. 

````python
plt.title('Corona Time Series in S.Korea') # 제목
plt.xlabel('Date') # x축 라벨
plt.ylabel('Cases') # y축 라벨
plt.show() 
````

+ 방법 2: 객체지향 API 이용

matplotlib 객체지향 API에는 `FigureCanvas`,`Renderer`,`Artist` 세 가지 객체로 구성되어 있다. 

- `FigureCanvas`: 그림을 그릴 영역을 나타내는 객체
- `Renderer`: 캔버스(`FigureCanvas`)에 그리는 도구 객체 
- `Artist`: `Renderer`가 `FigureCanvas`에 어떻게 그릴 것인지 나타내는 객체. 사용자 입장에서는 Artist 객체를 다루는데 집중하면 된다. `Artist`는 primitives 와 containers 라는 두가지 유형으로 구분할 수 있다. 
     + Primitives: `Line2D`,`Rectangle`,`Text`,`AxesImage`,`Patch`와 같이 캔버스에 그려지는 표준 그래픽 객체
     + Containers: `Axis`,`Axes`,`Figure`등과 같이 이들 primitives가 위치하게 될 대상 

객체 지향 방식으로 그림을 그리는 표준적인 방법은 다음과 같다 :-
1. 직접 `Figure` 객체를 생성한다.
2. 이를 이용해 하나 이상의 `Axes` 객체를 만든다.
3. `Axes`객체의 헬퍼함수로 primitives를 만든다. 

````python
fig = plt.figure() # 직접 Figure 객체 생성
ax= fig.subplots() # 직접 axes를 생성
ax.plot(covid_kor.index, covid_kor.cases) # 생성된 axes 에 대한 plot() 멤버 직접 호출 
````

+ 방법 3: 방법1과 방법2를 조합하여 `Figure`와 `Axes`를 `plt.subplots()` 라는 편의함수 사용한다. plt.subplots() 편의 함수는 Figure 객체를 생성하고 Figure.subplots()를 호출하여 리턴

````python
fig,ax = plt.subplots() 
ax.plot(covid_kor.index, covid_kor.cases) 
````
