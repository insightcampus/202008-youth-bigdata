```python

from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
url= "https://www.yna.co.kr/view/AKR20200204081000017?section=it/it"
html= requests.get(url).text 
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
text= soup.find('div',class_="story-news article").text.replace('\n','').replace("\'","")
# p_tag= div.find_all('p')
# for i in p_tag:
#     content+=i.text
# content[:500]
print(text)
```

```python
type(text)
```

```python
#Okt (Twitter tokenizer 가 v0.5.0. 부터 Okt로 변경)
from konlpy.tag import Okt
okt=Okt()
okt_tokens= okt.morphs(content)
print(okt_tokens)
```

```python
#Okt
oktTag = []
for token in okt_tokens:
    oktTag+= okt.pos(token)
print(oktTag)
```

```python
from collections import Counter
Counter(oktTag).most_common()
```

```python
#불용어 처리
stopPos = ['Determiner','Adverb','Conjuctions','Josa','PreEomi','Eomi','Suffix',
          'Punctuation','Foreign','Alpha','Number','Unknown']

stopWord = []
word=[]
for tag in oktTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag[0])
print(word)
```
