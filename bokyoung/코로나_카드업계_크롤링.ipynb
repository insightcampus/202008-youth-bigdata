```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

df = pd.DataFrame(columns=("date","title","contents"))

url= "https://search.naver.com/search.naver?where=news&query=%EC%BD%94%EB%A1%9C%EB%82%98%EC%B9%B4%EB%93%9C%EC%82%AC&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=1&ds=&de=&docid=&nso=so%3Ar%2Cp%3A1w%2Ca%3Aall&mynews=0&refresh_start={}&related=0"

DATE=[]
CONTENT=[]
TITLE=[]

for i in range(1,4000,10):
    URL = url.format(i)
    resp = requests.get(URL)
    soup= BeautifulSoup(resp.content, 'html.parser')
    li_tag = soup.find('ul',class_='type01').find_all('li')
    for num,j in enumerate(li_tag):
        try:
            naver_news= j.find("dl").find("dd",{"class":"txt_inline"}).find("a").text
            if naver_news == "네이버뉴스":
                news_url = j.find("dl").find("dd",{"class":"txt_inline"}).find("a")['href']
                detail_resp = requests.get(news_url)
                detail_soup = BeautifulSoup(detail_resp.content,'html.parser')
                title = detail_soup.find("div",{"class":"content"}).find('h3').text.strip()
                date = detail_soup.find("span",{"class":"t11"}).text[:10]
                content = detail_soup.find("div",{"class":'_article_body_contents'}).text.strip()
                content = re.sub("function _flash_removeCallback() {}",'',content).strip()
                DATE.append(date)
                TITLE.append(title)
                CONTENT.append(content)


        except:
            pass
```

```python
import pandas as pd
df = pd.DataFrame()
df['date'] = DATE
df['title'] = TITLE
df['content'] = CONTENT

df
```
