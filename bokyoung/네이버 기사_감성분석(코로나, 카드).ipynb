```python
import pandas as pd
import numpy as np
df = pd.DataFrame(pd.read_csv(r'C:\Users\user\OneDrive\Desktop\git_folder\202008-youth-bigdata\bokyoung\naver_news.csv'))
df
```

```python
import re 
# '한글'을 제외한 다른 문자 모두 제거
remove_except_ko = re.compile(r"[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]")
def preprocess(text):
    text = re.sub(remove_except_ko,' ',text).strip()
    return text

df['contents'] = df['content'].map(lambda x : preprocess(x))
```

```python
from konlpy.tag import Mecab
mecab = Mecab()
# mecab 사용해 토큰화, 한글자 제거, stop_word 제거
stop_word = ['께서','에서','에게','으로','이랑','까지','부터','s',
             '데일리','뉴스','기자','오류','우회','위한','함수','추가',
            '습니다','확산','지난','거리','전망','코로나','카드']
stopPos = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JC','JX',
           'SF','SE','SS','SP','SSO','SSC','SC','SY','SH','SL','SN',
           'MM','MAG','MAJ']
def token_mecab(text):
    text = mecab.pos(text)
    text = [i for i in text if len(i[0])>1]
    text = [i for i in text if i[0] not in stop_word]
    text = [i[0] for i in text if i[1] not in stopPos]
    return text
```

```python
# 토큰화 + 토큰리스트 생성 
def make_tokens(df):
    df['tokens'] = ''
    tokens_list=[]
    for i, row in df.iterrows():
        if i%500==0:
            print(i,'/',len(df))
        token = token_mecab(df['contents'][i])
        df['tokens'][i] = ' '.join(token)
    return df

df = make_tokens(df)
```

```python
neg_dict = pd.DataFrame(pd.read_csv(r'C:\Users\user\Desktop\KnuSentiLex\neg_pol_word.txt', sep='\n', quoting=3))
neg_dict.head()
```

```python
tone_ls = []
for doc in df['tokens']:
    #print(doc.split())
    #words=[]
    #words.append(doc.split(' '))
    splitdoc = doc.split()
    pos=0
    neg=0
    for word in splitdoc:
        if word in list(pos_dict['word']):
            pos += 1
        if word in list(neg_dict['word']):
            neg += 1
    tone = (pos-neg)/(pos+neg)
    tone_ls.append(tone)
df['tone'] = tone_ls

df
```

```python
positive_report=0
negative_report=0

for tone in df['tone']:
    if tone>0:
        positive_report +=1
    if tone<0:
        negative_report +=1
print("number of positive reports is",positive_report)
print("number of negative reports is",negative_report)
```
