```python
import pandas as pd
import numpy as np
df = pd.DataFrame(pd.read_csv(r'C:\Users\user\OneDrive\Desktop\git_folder\202008-youth-bigdata\bokyoung\naver_news.csv'))
df
```

```python
import re 
# '한글'을 제외한 다른 문자 모두 제거
remove_except_ko = re.compile(r"[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]")
def preprocess(text):
    text = re.sub(remove_except_ko,' ',text).strip()
    return text

df['contents'] = df['content'].map(lambda x : preprocess(x))
```

```python
df.head()
```

```python
from konlpy.tag import Mecab
mecab = Mecab()
# mecab 사용해 토큰화, 한글자 제거, stop_word 제거
stop_word = ['께서','에서','이다','에게','으로','이랑','까지','부터','하다','s',
             '데일리','뉴스','기자','오류','우회','위한','함수','추가',
            '습니다','확산','지난','거리','전망','코로나','카드']
stopPos = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JC','JX',
           'SF','SE','SS','SP','SSO','SSC','SC','SY','SH','SL','SN',
           'MM','MAG','MAJ']
def token_mecab(text):
    text = mecab.pos(text)
    text = [i for i in text if len(i[0])>1]
    text = [i for i in text if i[0] not in stop_word]
    text = [i[0] for i in text if i[1] not in stopPos]
    return text
```

```python
# 토큰화 + 토큰리스트 생성 
def make_tokens(df):
    df['tokens'] = ''
    tokens_list=[]
    for i, row in df.iterrows():
        if i%500==0:
            print(i,'/',len(df))
        token = token_mecab(df['contents'][i])
        df['tokens'][i] = ' '.join(token)
    return df

df = make_tokens(df)
```

```python
docs= list(df['tokens'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidv = TfidfVectorizer()
tfidv = tfidv.fit(docs)
tfidv.transform(docs).toarray()
vocab= tfidv.get_feature_names()
```

```python
tfidv.transform(docs).toarray()[0]
```

```python
import pandas as pd
df = pd.DataFrame(tfidv.transform(docs).toarray(), columns = vocab)
df
```

```python
# sum tfidf frequency of each term through documents
sums = df.sum(axis=0)
sums
```
