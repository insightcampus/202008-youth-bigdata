```python
import requests
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as bs
```

```python
url="https://www.yna.co.kr/view/AKR20200204081000017?section=it/it"
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
```

```python
all = soup.find("div", {"class" : "wrap-container"})
```

```python
p = all.find_all("p") 
```

```python
content=''
for i in p:
    content+=i.text #누적합: content 안에 이 값들을 넣어주세요
content
```

```python
import MeCab
m=MeCab.Tagger()
out=m.parse("MeCab이 잘 설치 되었는지 확인 중입니다.")
print(out)
```

```python
out=m.parse(content)
print(out)
```

```python
from konlpy.tag import Mecab
m=Mecab()
mecab_tokens=m.morphs(content)
print(mecab_tokens)
```

```python
MecabTag=[]
for token in mecab_tokens:
    MecabTag+=m.pos(token)
print(MecabTag)
```

```python
from collections import Counter
Counter(MecabTag).most_common()
```

```python
# 관형사, 부사, 접속사, 조사, 전어미, 어미, 접미사, 구둣점, 외국어("SL"), 알파벳, 숫자, 모르는 거 빼기
stopPos=["JKS","JKC","JKG","JKO","JKB","JKV","JKQ","JC","JX","EP","EF","EC","ETN","ETM","XPN","XSN","XSV","XSA"
        "XR","SF","SE","SSO","SSC","SC","SY","SH","SN"]
stopWord=["무단","전재","재","배포","금지"]
word=[]

for tag in MecabTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag[0])
print(word)
```
