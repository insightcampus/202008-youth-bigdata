# TF-IDF 직접구현
- TF 계산 :TF stands for Term Frequency; how many terms are in a document
- IDF 계산: 단어가 얼마나 많은 문서에 등장했는지를 나타내는 DF의 역수
- TF-IDF 계산: 둘이 곱하세요

```python
docs=['오늘 동물원에서 원숭이와 코끼리를 봤어',
     "동물원에서 원숭이에게 바나나를 줬어 바나나를"]
```

```python
doc_ls=[]
for doc in docs:
    doc_ls.append(doc.split())
doc_ls
```

```python
from collections import defaultdict
word2id=defaultdict(lambda: len(word2id))
word2id
```

```python
for doc in doc_ls:
    for token in doc:
        word2id[token]
word2id
```

```python
import numpy as np
BoW_ls=[]
for i, doc in enumerate(doc_ls):
    bow=np.zeros(len(word2id),dtype=int)
    print(bow)
    for token in doc:
        bow[word2id[token]]+=1 #해당토큰의 위치
    BoW_ls.append(bow.tolist())
#1차원 배열
BoW_ls
```

```python
from IPython.core import display as ICD
import pandas as pd
sorted_vocab=sorted((value,key)for key, value in word2id.items())
print('sorted_vocab',sorted_vocab)
```

```python
vocab=[]
for v in sorted_vocab:
    vocab.append(v[1])
print("vocab",vocab)
```

```python
for i in range(len(docs)):
    print("문서{} : {}".format(i,docs[i]))
    ICD.display(pd.DataFrame(BoW_ls,columns=vocab).T)
    print("\n\n")
```

```python
df1=pd.DataFrame(BoW_ls, columns=word2id.keys()).T;df1
mapping={0:"문서0-토큰빈도",1:"문서1-토큰빈도"}
df1 = df1.rename(columns=mapping)
```

```python
df1["문서0 내 전체 토큰빈도"]=len(doc_ls[0]);df1
```

```python
df1["문서1내 전체 토큰빈도"]=len(doc_ls[1]);df1
```

```python
df1["TF0"]=df1["문서0-토큰빈도"]/df1["문서0 내 전체 토큰빈도"];df1
```

```python
df1["TF1"]=df1["문서1-토큰빈도"]/df1["문서1내 전체 토큰빈도"];df1  
```

- 토큰이 등장한 문서 수: 해당 토큰이 해당 문서에 존재한다면 1을 더하라: 중복은 알아서 걸러질까?

```python
# 토큰이 등장한 문서 수
num_doc_sample=[]
num=np.zeros(len(vocab), dtype=int)
```

```python
if (vocab[0] in doc_ls[0])==True:
    num[0]+=1
print(num)
```

```python
num_docs=np.zeros(len(vocab), dtype=int)
for i in range(0,len(vocab)):
    if (vocab[i] in doc_ls[0])==True:
        num_docs[i]=num_docs[i]+1
    if (vocab[i] in doc_ls[1])==True:
        num_docs[i]=num_docs[i]+1
num_docs
```

```python
df1["토큰이 등장한 문서 수"]=num_docs;df1
```

```python
df1["IDF"]=np.log10(df1["문서 수"]/df1["토큰이 등장한 문서 수"])
df1["로그전"]=df1["문서 수"]/df1["토큰이 등장한 문서 수"]
df1
```

```python
df1["TF0TDF"]=(df1["TF0"]*df1["IDF"])
df1["TF1TDF"]=(df1["TF1"]*df1["IDF"])
df1
```
