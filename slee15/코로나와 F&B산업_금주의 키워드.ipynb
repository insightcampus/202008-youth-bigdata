```python
import numpy as np 
import pandas as pd 
```

```python
import requests 
from bs4 import BeautifulSoup as bs 
import re 
# url 쓸 때 start있는 부분을 중괄호로{} 쓸기 instead of number 
# query를 '코로나' 말고 '네이버'로 바꾸면 검색이다르게 됨! 
url = 'https://search.naver.com/search.naver?&where=news&query={}&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=1&ds=2020.09.03&de=2020.09.11&docid=&nso=so:r,p:1w,a:all&mynews=0&cluster_rank=43&start={}&refresh_start=0'
keyword = '코로나식품산업'
DATE = [] # 각 컬럼 나중에 df로 만들어주기 위해 빈 리스트 만들어주기 (입력하는 위치 유념!!)
TITLE = []
SOURCE = []
CONTENT = []
URL = []
#keyword는 띄어쓰기하면 검색안됨
for i in range(1,100, 10): # 1부터 30까지 10씩 증가하도록 페이지를 돌아라.. 라고 만들어 놓은거에여~
    URL = url.format(keyword, i) #저 위 사이트의 {}에다 keyword와 i range 넣기 
    resp = requests.get(URL)
    print(URL)
    soup = bs(resp.content, 'html.parser')
    li_tag = soup.find('ul', class_ = 'type01').find_all('li')
    for j in li_tag:
        try: 
            naver_news = j.find('dl').find('dd', class_ = 'txt_inline').find('a').text
            if naver_news == '네이버뉴스':
                news_url = j.find('dl').find('dd', class_ = 'txt_inline').find('a')['href']
                press = j.find('span', class_ = '_sp_each_source').text.strip()
                # '#text'라고 되어 있는 레이블들은 class_ = '#text'라고 하지 않고 그 뒤에 text.strip()라고 입력하기 
                detail_resp = requests.get(news_url)
                detail_soup = bs(detail_resp.content, 'html.parser')
                #detail_resp, detail_soup은 네이버뉴스로 가는 링크임! 그 네이버뉴스 링크 다라들어가서 형식/레이블들로 입력해야함!! 
                date = detail_soup.find('span', class_ = 't11').text[:10]
                title = detail_soup.find('div', class_ = 'content').find('h3').text.strip() # strip()넣으면 출력할때'/'같은 걸 없애줘서 더 깔끔하게 해줌
                content = detail_soup.find('div', class_ = '_article_body_contents').text.strip()
                content = re.sub('\/\/ flash 오류를 우회하기 위한 함수 추가\nfunction _flash_removeCallback\(\) \{\}','',content).strip()
                # 정규 표현식 (import re)로 불필요한 문구 제거 
                DATE.append(date) # 아까 위에 만든 빈 리스트에 값 append 해주기 (입력 위치 유념!!)
                TITLE.append(title)
                SOURCE.append(press)
                CONTENT.append(content)
                URL.append(URL)
        except: 
            pass 
df = pd.DataFrame() # 빈 데이터프레임 만들어주고 (입력 위치 유념!!)
df['date'] = DATE # 차례대로 컬럼 하나씩 만들어주기 
df['title'] = TITLE 
df['press'] = SOURCE
df['content'] = CONTENT 
df['url'] = URL 
```
