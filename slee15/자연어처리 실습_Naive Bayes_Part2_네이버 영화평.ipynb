## 2) 품사 부착(PoS Tagging)

```python
mecab_tag = [] 
for token in mecab_tokens: 
    mecab_tag += mecab.pos(token)
print(mecab_tag)
```

## 3) 불용어 처리 

```python
from collections import Counter #단어가 얼마나 나왔는지 빈도수 세줌 
Counter(mecab_tag).most_common()
```

```python
stopPos = ['Determiner', 'Adverb', 'Conjunction', 'Josa', 'PreEomi', 'Suffix', 'Punctuation', 'Foreign', 'Alpha', 'Number', 'Unknown']
stopWord = ['은', '는', '가', '건', '것', '이']

rev_final = []
for tag in mecab_tag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            rev_final.append(tag[0])
print(rev_final)
```

## 4) Naive Bayes로 예측모델 생성 

```python
import pandas as pd 
import numpy as np 
rev_final
```

```python
training_set = np.array(rev_final)
print(training_set)
```

```python
## 토큰 빈도수 및 문서별 토큰수 계산 
```

```python
# 토큰별 문서 내 빈도수 카운팅 
from collections import defaultdict 
wordfreq = defaultdict(lambda: [0,0]) # lambda : 새로운 단어가 추가될 때 default를 [0,0]으로 하겠다
# [긍정, 부정]
for doc, point in training_set: #긍정, 부정 두 가지로 나눴기 때문에 doc, point 두개로 for문 돌림 
    words = doc.split()
    for word in words:
        if point == 1:
            wordfreq[word][0] += 1 # 긍정인 경우 0번째 인덱스에 1을 더해라 
        else: 
            wordfreq[word][1] += 1 # 부정인 경우 1번째 인덱스에 1을 더해달라 
wordfreq
```
