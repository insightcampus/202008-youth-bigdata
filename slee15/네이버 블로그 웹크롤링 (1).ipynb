```python
url_list = []
title_list = []

total_page = 5
for i in tqdm(range(0, total_page)):
    i = i*5 + 1 
    url = "https://search.naver.com/search.naver?date_from={0}&date_option=0&date_to={1}&dup_remove=1&nso=&post_blogurl=&post_blogurl_without=&query={2}&sm=tab_pge&srchby=all&st=sim&where=post&start={3}".format(start_date, end_date, query_txt, i)
    print(url)
    driver.get(url)
    time.sleep(0.5)

    titles = "a.sh_blog_title._sp_each_url._sp_each_title"
    article_raw = driver.find_elements_by_css_selector(titles)

    for article in article_raw:
        url = article.get_attribute('href')
        url_list.append(url)

    for article in article_raw:
        title = article.get_attribute('title')
        title_list.append(title)

        print(title)
```

```python
df = pd.DataFrame({'url': url_list, 'title': title_list})

df.to_csv('blog_url.xlsx')
```

```python
url_load = pd.read_csv('blog_url.xlsx')

num_list = len(url_load)

print(num_list)
url_load
```

```python
def blog_crawling(page=1): 
    dict = {}

    number = 49 
    for i in tqdm(range(0, number)):
        url = url_load['url'][i]
        driver = webdriver.Chrome(r"chromedriver.exe")
        driver.get(url)

        try:
            driver.switch_to_frame('mainFrame')

            target_info = {}

            overlays = ".se-fs-.se-ff-"
            tit = driver.find_element_by_css_selector(overlays)
            title = tit.text
            title

            overlays = ".se_publishDate.pcol2"
            date = driver.find_element_by_css_selector(overlays)
            date_n_time= date.text

            overlays = ".se-component.se-text.se-l-default"
            contents = driver.find_elements_by_css_selector(overlays)

            content_list = []
            for content in contents:
                content_list.append(content.text)

            content_str = ' '.join(content_list)
            driver.close
            print(content_str)
            
        except Exception as e:    # 모든 예외의 에러 메시지를 출력할 때는 Exception을 사용
            print('예외가 발생했습니다.', e)
    
    dict = {"title": title, "date": date_n_time, "content": content_str}
```

```python
import platform 
import time
import datetime
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
import csv
```

```python
def save_data(blog_post):
    keys = blog_post[0].keys()
    
    with open('blog_crawling17.csv', 'w') as file: 
        writer = csv.DictWriter(file, keys)
        writer.writeheader()
        writer.writerows(blog_post)
        print('success save data')
```

```python
blog_post_list = []

for i in range(1, 50, 10): 
    blog_post_list.extend(blog_crawling(page=i))
    print("page = ", i)
    time.sleep(2)
    
save_data(blog_post_list)
```
