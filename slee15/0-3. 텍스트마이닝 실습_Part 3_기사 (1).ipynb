## 텍스트마이닝 실습: 기사 
링크: https://www.yna.co.kr/view/AKR20200204081000017?section=it/it

```python
import requests
from bs4 import BeautifulSoup as bs
url = 'https://www.yna.co.kr/view/AKR20200204081000017?section=it/it'
response = requests.get(url)
soup = bs(response.text, 'html.parser')
print(soup)
soup.prettify()
```

```python
text = soup.find().text.replace('\n', '')
print(url)
print(text)
```

```python
#크롤링 다른 방법 
div = soup.find('div', class_ = 'story-news article')
p_tag = div.find_all('p')

AI = ''
for i in p_tag:
    AI += i.text
    AI[:300]
print(AI)
```

## 1) 추출한 기사 내용 토큰화 

```python
from konlpy.tag import Mecab 
m = Mecab()
```

```python
#string인지 확인 아니면 string()
type(AI) 
```

```python
# 메켑
from konlpy.tag import Mecab
mecab = Mecab()
mecab_tokens = mecab.nouns(AI)
print(mecab_tokens)
```

```python
# 코모란
from konlpy.tag import Komoran 
komoran = Komoran()
komoran_tokens = komoran.nouns(AI)
print(komoran_tokens)
```

```python
#한나눔 
from konlpy.tag import Hannanum
hannanum = Hannanum()
hannanum_tokens = hannanum.nouns(AI)
print(hannanum_tokens)
```

```python
# Okt
from konlpy.tag import Okt
okt = Okt()
okt_tokens = okt.nouns(AI)
print(okt_tokens)
```

```python
# Kkma 
from konlpy.tag import Kkma 
kkma = Kkma()
kkma_tokens = kkma.nouns(AI)
print(kkma_tokens)
```

## 2) 품사 부착 (PoS Tagging) --> Okt로!

```python
okt_tag = [] 
for token in okt_tokens: 
    okt_tag += okt.pos(token)
print(okt_tag)
```

## 3) 불용어 처리 (Stopword)

```python
from collections import Counter #단어가 얼마나 나왔는지 빈도수 세줌 
Counter(okt_tag).most_common()
```

```python
stopPos = ['Determiner', 'Adverb', 'Conjunction', 'Josa', 'PreEomi', 'Suffix', 'Punctuation', 'Foreign', 'Alpha', 'Number', 'Unknown']
stopWord = ['은', '가', '건', '것', '이', '저작권자', '연합뉴스', '무단', '재' '배포']

AI_final = []
for tag in okt_tag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            AI_final.append(tag[0])
print(AI_final)
```
