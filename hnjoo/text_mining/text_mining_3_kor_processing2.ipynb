### 0. 필요한 패키지 다운 및 기사 크롤링

```python
import pandas as pd
import requests
# import re
from bs4 import BeautifulSoup
```

```python
url = 'https://www.yna.co.kr/view/AKR20200204081000017?section=it/it'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup) # 아래 결과들이 안나오면 크롤링 막은 홈페이지
```

```python
# 가져오는 첫 번째 방법
div = soup.find('div',class_='story-news article')
p_tag = div.find_all('p')
text = ''
for i in p_tag:
    text += i.text
text
```

```python
# 가져오는 두 번째 방법
text = soup.find('div', class_='story-news article').text.replace('\n','').replace("\'","") ; text
```

### 1. 토큰화(Tokenizing)

```python
# mecab 토큰화
from konlpy.tag import Mecab
# 선언
mecab = Mecab()
```

```python
mecab_tokens = mecab.morphs(text) # mecab.pos(content) 로 바로 해도 가능
```

```python
mecab_tokens[:20]
```

### 2. 품사 부착(PoS Tagging)
- 필수참고 : https://docs.google.com/spreadsheets/d/1OGAjUvalBuX-oZvZ_-9tEfYD2gQe7hTGsgUpiiBSXI8/edit#gid=0

```python
# mecab pos-tagging
mecabTag = []
for token in mecab_tokens:
    mecabTag += mecab.pos(token)
print(mecabTag[:20])
```

### 3. 불용어 처리(Stopword)
- 최빈어 조회 및 불용어 제거 대상 선정 > Counter
- 불용어 처리 > For loops

```python
# 최빈어 조회하여 불용어 제가 대상 선정
from collections import Counter
Counter(mecabTag).most_common()
```

```python
# 불용어 처리
stopPos = ['Determiner','Adverb','Conjunction','Josa','PreEomi','Eomi','Suffix',
          'Punctuation','Foreign','Alpha','Number','Unknown']
stopWord = ['을','다','는','은','(',')','=','저작권자','무단','전재','배포','금지','연합뉴스','로','<','>','.']
word = []
for tag in mecabTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag[0])
print(word)
```

```python
Counter(word).most_common()
```
