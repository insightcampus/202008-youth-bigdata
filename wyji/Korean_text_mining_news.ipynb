```
from konlpy.tag import Mecab
mecab = Mecab()
```

### 텍스트 데이터 가져오기

```
import requests
from bs4 import BeautifulSoup as bs
```

```
url = "https://www.yna.co.kr/view/AKR20200204081000017?section=it/it"
html=requests.get(url).text
soup=bs(html,'html.parser')
```

```
body = soup.find('div', class_='story-news article')
p_tag = body.find_all('p')
content = ''
for p in p_tag:
  content += p.text
content
```

### 텍스트 토큰화

```
from konlpy.tag import Mecab
mecab = Mecab()
mecab_tokens = mecab.morphs(content)
print(mecab_tokens)
```

### 품사부착

```
mecabTag = []
for token in mecab_tokens:
  mecabTag += mecab.pos(token)
mecabTag
```

### 불용어처리

```
stopPos = ['SSO','SSC','SL','SC','SY','SN','JKO','JKS','JKC','JKG','JKB','JKQ','JC','JKV','SF','JX','JX','VCP','VCN','MM','MAG','MAJ','VV+EP','MAG']
stopWord = ['가','일','은', '의']

word = []
for tag in mecabTag:
  if tag[1] not in stopPos:
    if tag[0] not in stopWord:
      word.append(tag[0])
print(word)
```
