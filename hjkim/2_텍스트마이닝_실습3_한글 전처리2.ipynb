```python
import pandas as pd
import requests
# import re
from bs4 import BeautifulSoup
```

```python
url = 'https://www.yna.co.kr/view/AKR20200204081000017?section=it/it'
resp = requests.get(url)
soup = BeautifulSoup(resp.text,'html.parser')
```

```python
text = soup.find('div',class_='article').text.replace('\n','')
print(url)
print(text)
```

```python
type(text)
```

```python
# mecab 토큰화
from konlpy.tag import Mecab
# 선언 
mecab = Mecab()
mecab_tokens = mecab.morphs(text)
print(mecab_tokens[:30])
```

```python
# mecab pos-tagging
mecabTag = []
for token in mecab_tokens:
    mecabTag += mecab.pos(token)
print(mecabTag[:30])
```

```python
#mecab
# 최빈어 조회. 최빈어를 조회하여 불용어 제거 대상을 선정
from collections import Counter
Counter(mecabTag).most_common()
```

```python
#불용어 처리
stopPos = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JC','JX',
           'SF','SE','SS','SP','SSO','SSC','SC','SY','SH','SL','SN',
           'MM','MAG','MAJ']
stopWord = ['을','다','는', '저작권자', '연합뉴스', '무단', '전재', '재', '배포', '금지', '송고']
word = []
for tag in mecabTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
             word.append(tag[0])         
print(word)
```
