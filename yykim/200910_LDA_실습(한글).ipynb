## 데이터전처리

```
exam['clean_doc'] = exam['content'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]","")

stopwords = ['전','재','(,)','수','더','"','-','와','보다','>','<','/','단','고',"했다",".","'",'가','며',",",'kr<저작권가','서울','=','연합뉴스','홍','만','과','지인','기자','로','ㄴ다','어','의','송','은','적','을','하','ㄴ','ㅁ','ㄹ','c','등','도','무단','전재','저작권','배포','금지','ljungberg','이날','되','게','yna','.co','.kr']
tokenized_docs_temp = []
for sentence in exam['clean_doc']:
    temp_doc = []
    temp_doc = okt.nouns(sentence)
    temp_doc = [word for word in temp_doc if not word in stopwords]
    tokenized_docs_temp.append(temp_doc)
```

```
import itertools
tokenized_docs_fin = list(itertools.chain(*tokenized_docs_temp))
```

```
type(tokenized_docs_fin)
```

## LDA 토픽모델링

```
def my_tokenizer(text):
    return text.split()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

tfidf_vect = TfidfVectorizer(tokenizer = my_tokenizer)
tfidf = tfidf_vect.fit_transform(tokenized_docs_fin)
lda = LatentDirichletAllocation(n_components=10, learning_method='online')

lda_output = lda.fit_transform(tfidf)
```
